```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.catplot(
    data=data,
    x='tip',
    y='size'
)
```

```python
sns.catplot(
    data=data,
    x='tip',
    y='size',
    hue='sex'
)
```

```python
sns.catplot(
    data=data,
    x='tip',
    y='size',
    hue='sex',
    palette='icefire'
)
```

```python
sns.catplot(
    data=data,
    x='tip',
    y='size',
    hue='sex',
    height=10
)
```

```python
sns.catplot(
    data=data,
    x='tip',
    y='size',
    hue='sex',
    kind='bar'
)
```

```python
sns.catplot(
    data=data,
    x='tip',
    y='size',
    hue='sex',
    kind='boxen'
)
```
