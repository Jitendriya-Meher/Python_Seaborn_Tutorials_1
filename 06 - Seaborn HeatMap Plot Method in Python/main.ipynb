Heat Map

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = np.linspace(1,10,20)
data
```

```python
data = data.reshape(4,5)
data
```

```python
sns.heatmap(data)
```

```python
data = sns.load_dataset('anagrams')
data.head()
```

```python
data = data.drop(columns=['attnr'])
data = data.head()
data
```

```python
sns.heatmap(data)
```

```python
sns.heatmap(
    data=data,
    vmin=1,
    vmax=20
)
```

```python
sns.heatmap(
    data=data,
    cmap='gist_heat'
)
```

```python
sns.heatmap(
    data=data,
    cmap='Blues'
)
```

```python
sns.heatmap(
    data=data,
    annot=True
)
```

```python
names = np.array([['a0','a1','a3','a4'],['a0','a1','a3','a4'],['a0','a1','a3','a4'],['a0','a1','a3','a4'],['a0','a1','a3','a4']])
sns.heatmap(
    data=data,
    annot=names,
    fmt='s'
)
```

```python
feat = {
    'fontsize':20,
    'color':'y'
}
sns.heatmap(
    data=data,
    annot=True,
    annot_kws=feat
)
```

```python
feat = {
    'fontsize':20,
    'color':'y'
}
sns.heatmap(
    data=data,
    annot=True,
    annot_kws=feat,
    linewidths=4,
    linecolor="g"
)
```

```python
feat = {
    'fontsize':20,
    'color':'y'
}
sns.heatmap(
    data=data,
    annot=True,
    annot_kws=feat,
    linewidths=4,
    linecolor="g",
    cbar=False
)
```

```python
feat = {
    'fontsize':20,
    'color':'y'
}
mp = sns.heatmap(
    data=data,
    annot=True,
    annot_kws=feat,
    linewidths=4,
    linecolor="g",
    cbar=False,
    xticklabels=False,
    yticklabels=False
)
mp.set(xlabel="python",ylabel="amit")
sns.set(font_scale=5)
```
