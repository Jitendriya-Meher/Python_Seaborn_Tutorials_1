pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.pairplot(
    data=data
)
```

```python
sns.pairplot(
    data=data,
    hue='sex'
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    vars=['size','tip']
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    vars=['size','tip'],
    hue_order=['Female','Male']
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    vars=['size','tip'],
    hue_order=['Female','Male'],
    palette='icefire'
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    hue_order=['Female','Male'],
    kind="reg"
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    hue_order=['Female','Male'],
    kind="kde"
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    hue_order=['Female','Male'],
    kind="hist"
)
```

```python
sns.pairplot(
    data=data,
    hue='sex',
    hue_order=['Female','Male'],
    markers=['*','>']
)
```
