Box plot

```python
import matplotlib.pylab as plt
import seaborn as sns
import  pandas as pd
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.set(style='whitegrid')
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='time'
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex'
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    color='r'
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    order=['Fri','Sat','Thur','Sun']
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    order=['Fri','Sat','Thur','Sun'],
    showmeans=True
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    order=['Fri','Sat','Thur','Sun'],
    showmeans=True,
    linewidth=3
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    order=['Fri','Sat','Thur','Sun'],
    showmeans=True,
    palette='plasma'
)
```

```python
sns.boxplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    order=['Fri','Sat','Thur','Sun'],
    showmeans=True,
    palette='plasma',
    orient='v'
)
```

```python
sns.boxplot(
    x='total_bill',
    y='day',
    data=data,
    hue='sex',
    order=['Fri','Sat','Thur','Sun'],
    showmeans=True,
    palette='plasma',
    orient='h'
)
```
