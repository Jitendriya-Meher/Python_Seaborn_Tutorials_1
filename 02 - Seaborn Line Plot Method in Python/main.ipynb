Line plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7,8,9]
var1 = [2,3,40,5,30,78,9,9,10]
```

```python
plt.plot(var,var1)
```

```python
sns.lineplot(x=var,y=var1)
```

```python
data = sns.load_dataset('penguins')
data
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=data)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex")
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex", size=10)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style='sex')
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",palette='Accent')
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style='sex',palette='Accent',markers=["o","*"])
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style='sex',palette='Accent',markers=["o","*"], legend=False)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style='sex',palette='Accent',markers=["o","*"], legend="full")
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",style='sex',palette='Accent',markers=["o","*"], legend="full")
plt.grid()
plt.title("python")
```
