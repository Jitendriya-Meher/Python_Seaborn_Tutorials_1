```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.boxplot(
    data=data,
    y='total_bill'
)
```

```python
sns.boxplot(
    data=data,
    y='total_bill',
)
plt.show()
```

```python
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
plt.show()
```

```python
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
plt.grid()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
plt.show()
```

```python
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(3,12))
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
sns.despine()
plt.show()
```

```python
sns.set_context(
    'poster',
    font_scale=1.2,
)
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
sns.despine()
plt.show()
```

```python
sns.set_context(
    'paper',
    font_scale=1.2,
)
sns.barplot(
    data=data,
    y='total_bill',
    x='day'
)
sns.despine()
plt.show()
```

```python
sns.set_context(
    'poster',
    font_scale=1.2,
)
sns.barplot(
    data=data,
    y='total_bill',
    x='day',
    palette='cool'
)
sns.despine()
plt.show()
```
