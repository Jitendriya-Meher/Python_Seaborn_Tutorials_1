scatterplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset('penguins')
data
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data
)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex'
)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex'
)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex',
    size='sex',
    sizes=(20,50)
)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex',
    size='sex',
    sizes=(20,50),
    palette='icefire'
)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex',
    size='sex',
    sizes=(20,50),
    palette='spring'

)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex',
    size='sex',
    sizes=(20,50),
    palette='icefire'
)
```

```python
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex',
    size='sex',
    sizes=(20,50),
    palette='icefire',
    alpha=0.5
)
```

```python
m = {
    "Male":"*",
    "Female":"o"
}
sns.scatterplot(
    x='bill_length_mm',
    y='bill_depth_mm',
    data=data,
    hue='sex',
    style='sex',
    size='sex',
    sizes=(100,60),
    palette='icefire',
    alpha=0.7,
    markers=m
)
```
