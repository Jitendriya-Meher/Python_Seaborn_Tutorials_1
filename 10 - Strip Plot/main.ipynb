strip plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.stripplot(
    x='day',
    y='total_bill',
    data=data
)
```

```python
sns.stripplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex'
)
```

```python
sns.stripplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    palette='ocean'
)
```

```python
sns.stripplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    palette='ocean',
    linewidth=1
)
```

```python
sns.stripplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    palette='ocean',
    linewidth=1,
    edgecolor="red"
)
```

```python
sns.stripplot(
    x='day',
    y='total_bill',
    data=data,
    hue='sex',
    palette='ocean',
    linewidth=1,
    edgecolor="red",
    jitter=1
)
```

```python
sns.stripplot(
    x=data['total_bill']
)
```
