Bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset('penguins')
data
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data)
```

```python
sns.barplot(x=data.island, y=data.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex')
```

```python
order = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', order=order)
```

```python
order = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', order=order,hue_order=["Female","Male"] )
```

```python
order = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', order=order, errorbar=('ci', 10))
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', orient='h')
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=data, hue='sex', orient='v')
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=data, hue='sex', orient='h')
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, color='g')
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='Accent')
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire')
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire', saturation=10)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire', errcolor='y')
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire', errcolor='y', errwidth=10)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire', errcolor='b', capsize=0.2)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire', errcolor='b', capsize=0.2, dodge=False)
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data, hue='sex', palette='icefire', capsize=0.2, alpha=0.7)
```
