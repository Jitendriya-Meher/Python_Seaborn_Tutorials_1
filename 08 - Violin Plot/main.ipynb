violine plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    hue="time"
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    linewidth=5
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    linewidth=3
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    palette='icefire'
)
```

```python
sns.violinplot(
    x = 'time',
    y='total_bill',
    data=data,
    palette='icefire'
)
```

```python
sns.violinplot(
    x = 'time',
    y='total_bill',
    data=data,
    palette='icefire',
    order=['Dinner','Lunch']
)
```

```python
sns.violinplot(
    x = 'time',
    y='total_bill',
    data=data,
    palette='icefire',
    saturation=0.4
)
```

```python
sns.violinplot(
    x = 'time',
    y='total_bill',
    data=data,
    color='g'
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    palette='icefire',
    hue='sex'
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    palette='icefire',
    hue='sex',
    split=True
)
```

```python
sns.violinplot(
    x = 'day',
    y='total_bill',
    data=data,
    palette='icefire',
    hue='sex',
    scale='width'
)
```

```python
sns.violinplot(
    y = 'day',
    x='total_bill',
    data=data,
    palette='icefire',
    hue='sex'
)
```

```python
sns.violinplot(
    x = 'time',
    y='total_bill',
    data=data,
    palette='icefire',
    inner='stick'
)
```

```python
sns.violinplot(
    data['total_bill']
)
```
