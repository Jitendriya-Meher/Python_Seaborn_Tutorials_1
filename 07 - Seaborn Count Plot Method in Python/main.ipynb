Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('tips')
data
```

```python
sns.countplot(
    x='sex',
    data=data
)
```

```python
sns.countplot(
    x='smoker',
    data=data
)
```

```python
sns.barplot(
    x="sex",
    y="size",
    data=data
)
```

```python
sns.countplot(
    x='sex',
    hue='smoker',
    data=data
)
```

```python
sns.countplot(
    y='sex',
    hue='smoker',
    data=data
)
```

```python
sns.countplot(
    x='sex',
    hue='smoker',
    data=data,
    palette='icefire'
)
```
